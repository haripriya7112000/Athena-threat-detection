```python
# Day 3: Data Preprocessing

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import os

# Load dataset
DATA_PATH = '../data/CICIDS2017/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv'
df = pd.read_csv(DATA_PATH)

# Preview
print("Dataset shape:", df.shape)
df.head()
```

```python
# Clean column names
print("Original columns:", df.columns.tolist()[:76])
df.columns = df.columns.str.strip()
print("Cleaned columns:", df.columns.tolist()[:76])
```

```python
# Strip whitespace from column names
df.columns = df.columns.str.strip()

# Replace infinite values with NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Drop rows with NaN
df.dropna(inplace=True)

# Select only numeric columns for scaling
features = df.select_dtypes(include=[np.number]).columns
```

```python
# Scale numeric features
scaler = StandardScaler()
df[features] = scaler.fit_transform(df[features])

# Save the processed dataset
os.makedirs('../data/processed', exist_ok=True)
df.to_csv('../data/processed/processed_ddos.csv', index=False)

print("Preprocessing complete. Data saved to '../data/processed/processed_ddos.csv'")
```
